```python
#leer el archivo
data = pd.read_csv("Ventas_totales.csv")
data.head()
#data.info()
```

```python
# Identificar valores nulos
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
data1 = data.copy()
```

```python
# Sustitución por media
data1['subtotal_ventas_alimentos_bebidas'] = data1['subtotal_ventas_alimentos_bebidas'].fillna(round(data1['subtotal_ventas_alimentos_bebidas'].mean(),1))
data1['salon_ventas'] = data1['salon_ventas'].fillna(round(data1['salon_ventas'].mean(),1))
data1['almacen'] = data1['almacen'].fillna(round(data1['almacen'].mean(),1))
data1['otros'] = data1['otros'].fillna(round(data1['otros'].mean(),1))
# ocupamos promedio al ser una de las mejores maneras de sustitución de valores nulos, sobre todo para operaciones o el total de ventas

# también ocupamos este método por el tamaño de los números, decidimos ocuparlo
data1['panaderia'] = data1['panaderia'].fillna(round(data1['panaderia'].mean(),1))
data1['verduleria_fruteria'] = data1['verduleria_fruteria'].fillna(round(data1['verduleria_fruteria'].mean(),1))
```

```python
# Sustitución por mediana
data1['indumentaria_calzado_textiles_hogar'] = data1['indumentaria_calzado_textiles_hogar'].fillna(round(data1['indumentaria_calzado_textiles_hogar'].median(),1))
data1['electronicos_articulos_hogar'] = data1['electronicos_articulos_hogar'].fillna(round(data1['electronicos_articulos_hogar'].median(),1))

#ocupamos la mediana porque también funciona bien como método de sustitución, se ocupó además en las variables que tienen más de un nulo
```

```python
#Método por ffill
data1['tarjetas_debito'] = data1['tarjetas_debito'].fillna(method='ffill')
data1['tarjetas_credito'] = data1['tarjetas_credito'].fillna(method='ffill')
data1['otros_medios'] = data1['otros_medios'].fillna(method='ffill')
```

```python
# Método bfill
data1['bebidas'] = data1['bebidas'].fillna(method='bfill')
data1['lacteos'] = data1['lacteos'].fillna(method='bfill')
data1['carnes'] = data1['carnes'].fillna(method='bfill')
data1['alimentos_preparados_rotiseria'] = data1['alimentos_preparados_rotiseria'].fillna(method='bfill')
```

Se ocuparon los métodos de forward y backfills a variables con solo un nulo, ya que de esta manera no se afecta al dataset

```python
#Corroboramos que ya no existan valores nulos
valores_nulos = data1.isnull().sum().sum()
valores_nulos
```

```python
#convertir datagrame a csv
data1.to_csv("VentasTotales_limpio.csv")
```
